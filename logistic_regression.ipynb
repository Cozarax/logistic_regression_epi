```python
import numpy as np
import import_ipynb 
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import datasets
from tqdm import tqdm

from neuronPOO import artificial_neuron
```

```python
scaler = StandardScaler()
bc = datasets.load_breast_cancer()
X, y = bc.data, bc.target.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

X_train_reshape = scaler.fit_transform(X_train)
X_test_reshape = scaler.transform(X_test)

clf = artificial_neuron(n_iter = 1000, learning_rate = 0.01) 
clf.fit(X_train_reshape, y_train)

y_pred = clf.predict(X_test_reshape).flatten()

def accuracy(y_pred, y_test):
  return np.sum(y_pred == y_test.flatten())/len(y_test)

acc= accuracy(y_pred, y_test)
print(acc)
```
